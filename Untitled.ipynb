```python
# Importing libraries
import pandas as pd
from pandas import Series
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
%matplotlib inline
```

```python
df = pd.read_csv('zillow_data.csv')
df
```

```python
# Calculate the ROI column
df['ROI'] = round((df['2018-04'] - df['1996-04']) / df['1996-04'], 4)

# Display the DataFrame with the new 'ROI' column
print(df)
```

```python
def melt_data(df):
    melted = pd.melt(df, id_vars=['RegionID','RegionName','CountyName','City','State','Metro','SizeRank','ROI'], 
                     var_name='Date')
    melted['Date'] = pd.to_datetime(melted['Date'])
    melted = melted.dropna(subset=['value'])
    return melted

melted_df = melt_data(df)
melted_df
```

```python
#Change Zipcode dtype to 'str'
melted_df['RegionID'] = melted_df['RegionID'].astype(str)

# Make sure the data type of the 'Date' column is datetime
melted_df['Date'] = pd.to_datetime(melted_df['Date'], format='%m/%y')

# Set the 'Date' column as index
melted_df.set_index('Date', inplace=True)
```

```python
melted_df = melted_df['2008-01-01':'2018-04-01']
melted_df
```

```python
# Count non-numeric values in the 'value' column
non_numeric_count = melted_df['value'].apply(lambda x: pd.to_numeric(x, errors='coerce')).isna().sum()

print(f"Number of non-numeric values in 'value' column: {non_numeric_count}")
```

```python
melted_df['value'] = melted_df['value'].astype(str).str.strip()

```

```python
melted_df.resample('MS').mean().index
```

```python
# Drop rows where 'value' is not a float or int
melted_df = melted_df[pd.to_numeric(melted_df['value'], errors='coerce').notna()]

# Optional: Convert 'value' column to float if needed
melted_df['value'] = melted_df['value'].astype(float)
```
