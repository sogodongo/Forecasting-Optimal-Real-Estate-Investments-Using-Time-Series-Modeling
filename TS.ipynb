```python
import pandas as pd

# Load the Zillow dataset
zillow_data = pd.read_csv('zillow_data.csv')

# Define the date columns
date_columns = zillow_data.columns[7:]

# Convert the numerical date values to strings (YYYY-MM format)
zillow_data[date_columns] = zillow_data[date_columns].astype(str)

# Define a function to clean and format the date values
def clean_and_format_date(date_str):
    # Split the string on "." and keep the first part
    date_parts = date_str.split(".")
    if len(date_parts) > 0:
        # Extract the first 7 characters (YYYY-MM)
        return date_parts[0][:7]
    else:
        return date_str  # If the format is already correct, return it as is

# Apply the clean_and_format_date function to the date columns
zillow_data[date_columns] = zillow_data[date_columns].applymap(clean_and_format_date)

# Now, convert the cleaned date strings to datetime objects
zillow_data[date_columns] = zillow_data[date_columns].apply(lambda x: pd.to_datetime(x, format='%Y-%m'), axis=0)

# Verify that the conversion was successful
print(zillow_data[date_columns].dtypes)


```
