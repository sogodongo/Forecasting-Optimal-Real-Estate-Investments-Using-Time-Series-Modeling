## 3. Loading and Filtering Data

```python
# Load the Zillow dataset
zillow_data = pd.read_csv('zillow_data.csv')
zillow_data
```

### ZIP Code Selection

The client is open to real estate investments anywhere in the country, indicating a lack of specific preferences. This approach allows for a comprehensive analysis, considering a wide range of opportunities and maximizing returns.

Thus I  decided  to use evaluation criteria from prior research enhancing the analysis by leveraging existing knowledge, saving time, and increasing credibility. This approach ensured a well-informed and reliable evaluation process. The three evaluation metrics chosen were:

**a. Historical Price Appreciation:**
   - Calculating historical price appreciation is essential for evaluating real estate investments. It offers insights into ZIP code performance over time, considering long-term trends, risk assessment, and profit potential.

**b. SizeRank:**
   - SizeRank measures the relative size or popularity of ZIP codes. It's valuable for assessing market significance, diversification opportunities, and growth potential in real estate investments.

**c. Coefficient of Variation (CV) for each ZIP code:**
   - The CV assesses data volatility, helping in risk assessment and investment strategy. It ensures portfolio diversity by considering different market stability levels.

**d. Return on Investment(ROI)**
   - In our real estate investment analysis, the strategic use of ROI stands as a pivotal metric. ROI quantifies the profitability of each ZIP code investment by assessing the change in median housing prices from April 1996 to April 2018. This strategic choice empowers us to gauge the performance of different ZIP codes, distinguishing those with substantial property value appreciation. By employing ROI, we rank and identify investment opportunities, guiding our selection of ZIP codes for optimal returns. Thus, predicting ROI is a metric upon which our final investment opportunities' selection will be based.

**a. Historical Price Appreciation**

```python
# Select relevant columns for historical price appreciation calculation
price_columns = zillow_data.columns[7:]  # Columns from 1996-04 to 2018-04
zillow_data['Historical_Price_Appreciation'] = zillow_data[price_columns].pct_change(axis=1).mean(axis=1)

# Sort the ZIP codes by historical price appreciation in descending order
zillow_data.sort_values(by='Historical_Price_Appreciation', ascending=False).head(20)
```

**b. Size Rank**

Calculating the threshold for the top 5% of SizeRank values is a strategic approach to pinpoint markets with significant size and potential for real estate investments. It aligns with the goals of diversification, risk mitigation, and maximizing profit potential.

```python
# Calculate the threshold for the top 5% of SizeRank values
size_rank_threshold = zillow_data['SizeRank'].quantile(0.95)
size_rank_threshold
```

Close to 13986 met this criteria

**c.  Coefficient of Variation (CV) for each ZIP code**

```python
# Calculate the coefficient of variation (CV) for each ZIP code
zillow_data['CV'] = zillow_data[price_columns].std(axis=1) / zillow_data[price_columns].mean(axis=1)

# Select the top 20 ZIP codes with the highest CV
top_20_cv_zipcodes = zillow_data.nlargest(20, 'CV')
top_20_cv_zipcodes
```

In our real estate investment analysis, we have strategically chosen to calculate the 60th percentile of the Coefficient of Variation (CV). This decision aligns with our risk assessment approach, catering to diverse investment strategies. By focusing on the 60th percentile, we strike a balance between stability and market variability, making it an attractive choice for risk-aware investors. Additionally, this approach promotes portfolio diversity, acknowledging the value of including markets with different levels of price stability. Ultimately, our goal is to target ZIP codes that offer a balanced combination of risk and profit potential, ensuring a well-rounded and strategic investment strategy.

```python
# Calculate the 60th percentile of CV
cv_threshold = zillow_data['CV'].quantile(0.6)
```

```python
# Filter ZIP codes to meet all criteria
filtered_zipcodes = zillow_data[
    (zillow_data['Historical_Price_Appreciation'] >= zillow_data['Historical_Price_Appreciation'].quantile(0.8)) &
    (zillow_data['SizeRank'] >= size_rank_threshold) &
    (zillow_data['CV'] < cv_threshold)
]
```

**d). Return on Investment (ROI)**

```python
# Drop rows with NaN values in the columns needed for ROI calculation
filtered_zipcodes = filtered_zipcodes.dropna(subset=['1996-04', '2018-04'])

# Calculate ROI for each ZIP code based on the change in median housing prices
filtered_zipcodes['ROI'] = (filtered_zipcodes['2018-04'] - filtered_zipcodes['1996-04']) / filtered_zipcodes['1996-04']

# Select the top 10 ZIP codes with the highest ROI
top_20_filtered_zipcodes = filtered_zipcodes.nlargest(20, 'ROI')

# Display the top 10 ZIP codes that meet all criteria
print("Top 10 ZIP Codes Based on Combined Criteria:")
top_20_filtered_zipcodes[['RegionName', 'Historical_Price_Appreciation','CV', 'SizeRank', 'ROI']]
```

## 4. Data Preprocessing

In this crucial data preprocessing step, our primary focus was to thoroughly examine the dataset and proactively tackle potential data issues that could potentially hinder the effectiveness of our time series modeling efforts. We concentrated on three pivotal aspects: outlier detection, missing value handling, and relevance assessment. Firstly, we employed robust statistical methods to pinpoint and address outliers, recognizing their capacity to exert substantial influence on the accuracy and dependability of our time series models. Secondly, we placed a significant emphasis on handling missing data, a paramount concern in time series modeling, by implementing tailored strategies like imputation or removal, depending on the dataset's specific context. Lastly, we conducted a comprehensive evaluation of each dataset column, leading to the removal of non-relevant features that did not contribute substantially to our time series modeling objectives. This careful data curation process ensured the dataset's quality and suitability for our forthcoming focused modeling endeavors.


```python
def get_datetimes(top_20_filtered_zipcodes):
    """
    Takes a dataframe:
    returns only those column names that can be converted into datetime objects 
    as datetime objects.
    """
    date_columns = [col for col in top_20_filtered_zipcodes.columns[7:] if re.match(r'\d{4}-\d{2}', col)]  # Filter columns with "YYYY-MM" format
    return pd.to_datetime(date_columns, format='%Y-%m')

# Use the modified get_datetimes function to obtain datetime columns
datetime_columns = get_datetimes(top_20_filtered_zipcodes)
datetime_columns
```

```python
top_20_filtered_zipcodes.dropna(axis=1)
```

```python
# Function to identify outliers using Z-score
def handle_outliers(data, column_name, threshold=3):
    # Convert the column to numeric (if it's not already)
    data[column_name] = pd.to_numeric(data[column_name], errors='coerce')
    
    # Calculate Z-scores for numeric values
    z_scores = np.abs(stats.zscore(data[column_name].dropna()))
    
    # Create a mask for outliers
    is_outlier = z_scores > threshold
    
    # Create a boolean array with the same length as the original DataFrame
    # and use it to index the DataFrame to select rows with outliers
    outliers = pd.Series(False, index=data.index)
    outliers[data[column_name].notna()] = is_outlier
    
    # Return the rows with outliers
    return data[outliers]

# Dictionary to store outliers for each column
outliers_dict = {}

# Exclude the specified columns
exclude_columns = ['RegionID', 'RegionName', 'City', 'State', 'Metro', 'CountyName']

# Get the list of numeric columns excluding the specified columns
numeric_columns = [col for col in top_20_filtered_zipcodes.columns if col not in exclude_columns]

# Loop through numeric columns and identify outliers
for column_name in numeric_columns:
    outliers = handle_outliers(top_20_filtered_zipcodes, column_name)
    
    if not outliers.empty:
        outliers_dict[column_name] = outliers

# Print the rows with outliers for each numeric column
for column_name, outliers in outliers_dict.items():
    print(f"Outliers in '{column_name}':")
    print(outliers)
```

```python
# Check for missing values in the DataFrame
missing_values = top_20_filtered_zipcodes.isnull().sum()

# List columns with missing values (if any)
columns_with_missing_values = missing_values[missing_values > 0]

# Display the columns with missing values and the number of missing values
if not columns_with_missing_values.empty:
    print("Columns with missing values:")
    print(columns_with_missing_values)
else:
    print("No missing values found in the DataFrame.")
```

```python
# Define the columns to exclude
exclude_columns = ['RegionID', 'RegionName', 'City', 'State', 'Metro', 'CountyName']

# Exclude the specified columns
filtered_df = top_20_filtered_zipcodes.drop(columns=exclude_columns)

# Calculate the mean (average) values for each column
column_means = filtered_df.mean()

# Fill missing values with the respective column means
top_20_filtered_zipcodes_filled = top_20_filtered_zipcodes.copy()
top_20_filtered_zipcodes_filled[filtered_df.columns] = top_20_filtered_zipcodes[filtered_df.columns].fillna(column_means)
```

## 6.Time Series Modelling

### a) ZIP Code Selection

In the domain of time series modeling, we embark on a venture to utilize historical data for the purpose of making predictions about the future.. To initiate our Time Series model, our initial step is to pinpoint a specific zipcode within the country, where we aim to predict future ROI.Our strategy involves selecting a single zipcode to construct our Time Series model initially, and for this purpose, we've opted for the one exhibiting the most substantial growth since 2011, namely, ZipCode-38461. This chosen zipcode will serve as our starting point for forecasting its future value. Subsequently, we will employ a similar workflow on other selected zipcodes systematically identifying the top 5 zipcodes anticipated to yield the highest ROI.

```python
#Dataframe containining historical prices of the 38461 Zip Code.

zip38461 = top_20_filtered_zipcodes[top_20_filtered_zipcodes['RegionName']==38461]
zip38461.head()
```

```python
#Meltinga and reshaping a DataFrame while converting date columns to datetime format, removing rows with missing values in the 'value' column, and returning the resulting DataFrame.

def melt_data(zip38461):
    melted = pd.melt(zip38461 , id_vars=['RegionID','RegionName','CountyName','CV','Historical_Price_Appreciation','City','State','Metro','SizeRank','ROI'], 
                     var_name='Date')
    melted['Date'] = pd.to_datetime(melted['Date'])
    melted = melted.dropna(subset=['value'])
    return melted

melted_df = melt_data(zip38461)
melted_df
```

```python
# Convert the 'Date' column to a datetime type
melted_df['Date'] = pd.to_datetime(melted_df['Date'])

# Set the 'Date' column as the index
time_series_df = melted_df.set_index('Date')

# Select the 'value' column as the time series data
time_series_df = time_series_df[['value']]
time_series_df
```

```python
#Visualizing our Time Series Data for the zipcode

fig, ax = plt.subplots(figsize=(20,12))
ax.plot(time_series_df)

ax.set_xlabel('Year', fontsize=30)
ax.set_ylabel('Price (in 100,000($)',fontsize=20)
ax.set_title('Price History for Zip Code 38461 ',fontsize=30);
```

The pattern observed in the graph is quite similar to what we encountered in our previous exploration of other zip codes.


Its observed that the market conditions before the 2010s do not necessarily mirror the current market situation. Hence, I have decided to focus our Time Series modeling solely on data from 2011 onwards.

```python
recent_series = time_series_df['2011':]
```

```python
#Visualizing our Time Series Data for zipcode- 38461 from 2011

fig, ax = plt.subplots(figsize=(20,12))
ax.plot(recent_series)

ax.set_xlabel('Year', fontsize=20)
ax.set_ylabel('Price in 100,000($)',fontsize=20)
ax.set_title('Zip Code 38461 Price (since 2011)',fontsize=30);
```

Observation:There is  a favorable trend characterized by a moderately consistent linear increase in average prices.

### b) Modelling

```python
# Splitting the last 24 months of our series as a test dataset,representing approximately 29% of our data.
train_series = recent_series[:'2016-04']
test_series = recent_series['2016-05':]
```

```python
# Selecting the model

auto_model = pm.auto_arima(train_series, start_p=0, start_q=0,
                     test='adf',
                     max_p=5, max_q=5,
                     m=1,
                     d=0,
                     seasonal = True,
                     start_P=0,start_Q=0,
                     D=0,
                     trace=True,
                     error_action= 'ignore',
                     suppress_warnings=True,
                     stepwise=True,with_intercept=False)

print(auto_model.summary())
```

This code automatically selects the best ARIMA model for the time series and provides a summary of the selected model's parameters and statistics.I shall proceed to plug in the optimal parameter values precsibed

```python
# Plug the optimal parameter values into a new SARIMAX model
ARIMA_MODEL = sm.tsa.statespace.SARIMAX(train_series, 
                                        order=(3, 0, 2), 
                                        seasonal_order=(0, 0, 0, 0), 
                                        enforce_stationarity=False, 
                                        enforce_invertibility=False)

# Fit the model and print results
output = ARIMA_MODEL.fit()

print(output.summary().tables[1])
auto_model.plot_diagnostics(figsize=(18, 18))
plt.show()
```

Inspecting the diagnostic plots to ensure that our residuals exhibit no correlation and adhere to a normal distribution. If these assumptions are not met, we may need to refine our model.

a) _the residuals_ closely align with the standard normal distribution (N(0,1)), indicating normal distribution characteristics.

b) _The QQ-plot_ demonstrates that the residuals adhere to a linear trend line, confirming their normal distribution.

c) _The correlogram plot_ illustrates minimal correlations between time series residuals and their lagged counterparts. This suggests the absence of evident seasonality in our data.

Thus, these observations instill confidence that our model does not have correlations and effectively fits our data for future price forecasting.


### c) Testing Forecasting Model Performance


Obtaining the predictions generated by our model for the time spanning from May 2016 to April 2018, which aligns with the duration used for our test dataset. We can visualize these predictions on a shared chart to assess their proximity or divergence from one another.

```python
# Get predictions starting from 04-01-2015 and calculate confidence intervals
pred = output.get_prediction(start=pd.to_datetime('2015-05'), end=pd.to_datetime('2018-04'), dynamic=False)
pred_conf = pred.conf_int()
pred_conf
```

**Plotting this  against our test data**

```python
# Set the figure size using plt.rcParams
plt.rcParams['figure.figsize'] = (18, 8)

# Plot observed values
ax = recent_series.plot(label='observed')

# Plot predicted values
pred.predicted_mean.plot(ax=ax, label='Prediction Series', alpha=0.9)

# Plot the range for confidence intervals
ax.fill_between(pred_conf.index,
                pred_conf.iloc[:, 0],
                pred_conf.iloc[:, 1], color='lightpink', alpha=0.5, label='Confidence Interval')

# Set axes labels
ax.set_xlabel('Date', fontsize=20)
ax.set_ylabel('Price', fontsize=20)
ax.set_title('Testing Forecasting Model Performance', fontsize=30)
plt.legend()
plt.show()
```

While our model wasn't flawless in its accuracy, it did successfully capture the general trend that unfolded between May 2015 and April 2018. With confidence in its performance over this 2-3 year timeframe, we can now proceed to predict future trends until April 2023. This endpoint corresponds to a full three years from the purchase date in April 2018

## 7.Forecasting 


In this section, we delve into the critical aspect of forecasting. Leveraging the insights gained from our model's performance evaluation and its ability to capture historical trends, we now shift our focus toward predicting future developments and outcomes. The  objective is to provide forecasts and projections that extend beyond our current point in time to capture the best 5 ZIP Codes to invest in from our list of 10.

```python
# Plug the parameter values from our Auto ARIMA model into a new SARIMAX model that fits the entire series.
ARIMA_MODEL = sm.tsa.statespace.SARIMAX(recent_series, 
                                        order=(3,0,2), 
                                        seasonal_order=(0, 0, 0, 0), 
                                        enforce_stationarity=False, 
                                        enforce_invertibility=False)

# Fit the model and print results
full_output = ARIMA_MODEL.fit()

print(output.summary().tables[1])
```

```python
## Getting a forecast for the next 84 months after the last recorded date on our dataset.
forecast = full_output.get_forecast(84)
future_prediction = forecast.conf_int()
future_prediction['value'] = forecast.predicted_mean
future_prediction.columns = ['lower','upper','prediction'] 
future_prediction
```

Having generated an eight-year forecast, it's time to create a visual representation to gauge the projected growth of the 38461 zipcode region by April 2025

```python
#Plotting our Forecast

fig, ax = plt.subplots()
recent_series.plot(ax=ax,label='Real Values')


future_prediction['prediction'].plot(ax=ax,label='predicted value',ls='--')

ax.fill_between(x= future_prediction.index, y1= future_prediction['lower'], 
                y2= future_prediction['upper'],color='lightpink',
                label='Confidence Interval')
ax.legend() 
plt.ylabel("Average Price")
plt.title('Average Home Price - 38461 - With Forcasted Value & Confidence Intervals')
plt.show()
```

Now that we've outlined a methodology for predicting future prices in zipcode 38461, we can seamlessly integrate this entire process into a repetitive loop to obtain our anticipated prices in April 2025 for the other selected filtered zipcode

```python
def melt_data(top_20_filtered_zipcodes):
    melted = pd.melt(top_20_filtered_zipcodes , id_vars=['RegionID','RegionName','CountyName','CV','Historical_Price_Appreciation','City','State','Metro','SizeRank','ROI'], 
                     var_name='Date')
    melted['Date'] = pd.to_datetime(melted['Date'])
    melted = melted.dropna(subset=['value'])
    return melted

melted_df1 = melt_data(top_20_filtered_zipcodes)
melted_df1
```

```python
# Convert the 'Date' column to a datetime type
melted_df1['Date'] = pd.to_datetime(melted_df1['Date'])

# Set the 'Date' column as the index
time_series_df1 = melted_df1.set_index('Date')

# Select the 'value' column as the time series data
time_series_df1 = time_series_df1[['value']]
time_series_df1
```

```python
zip_predictions = {}

# Loop through the top 10 filtered zipcodes
for zipcode in top_20_filtered_zipcodes['RegionName']:
    
  # Melting the time series of our given zipcode
    series1 = melt_data(top_20_filtered_zipcodes[top_20_filtered_zipcodes['RegionName'] == zipcode])
    
    # Convert the 'Date' column to a datetime type and set it as the index
    series1['Date'] = pd.to_datetime(series1['Date'])
    series1.set_index('Date', inplace=True)
    
    # Only taking data from 2013 onwards to more accurately reflect current market conditions
    recent_series = series1['2013':]
    
    # Splitting the last 36 months of our series as a test dataset.
    train_series = recent_series[:'2017-04']
    test_series = recent_series['2017-05':]
    
    
    # Auto ARIMA model
    auto_model = pm.auto_arima(train_series['value'], start_p=0, start_q=0,
                     test='adf',
                     max_p=5, max_q=5,
                     m=1,
                     d=0,
                     seasonal=True,
                     start_P=0, start_Q=0,
                     D=0,
                     trace=True,
                     error_action='ignore',
                     suppress_warnings=True,
                     stepwise=True, with_intercept=False)
   
    # Plug the optimal parameter values for our Training data into a SARIMAX model that fits our entire series.
    ARIMA_MODEL = sm.tsa.statespace.SARIMAX(recent_series['value'], 
                                            order=auto_model.order, 
                                            seasonal_order=auto_model.seasonal_order, 
                                            enforce_stationarity=False, 
                                            enforce_invertibility=False)

    # Fit the model and print results
    output = ARIMA_MODEL.fit()

    # Getting a forecast for the next 84 months after the last recorded date on our dataset.
    forecast = output.get_forecast(steps=84)
    prediction = forecast.conf_int()
    prediction['value'] = forecast.predicted_mean
    prediction.columns = ['lower', 'upper', 'prediction'] 
    
    # Adding the Zipcode's ROI to the zip_predictions dictionary
    zip_predictions[zipcode] = ((prediction['prediction'][-1])
                                - (recent_series['value'][-1])) / (recent_series['value'][-1])
```

Predictions in financial forecasting are heavily reliant on historical monthly returns, forming the cornerstone of the analysis. While historical data can provide valuable insights and offer a glimpse into trends, it lacks the ability to account for unforeseen events or shifts in economic, social, and environmental conditions that could significantly impact the subject of the forecast.

Investments, economic conditions, and various other factors are inherently subjected to uncertainties and risks that often elude the purview of historical data alone. These uncertainties may encompass abrupt changes in government policies, rapid technological advancements, fluctuations in market sentiment, and the influence of global events, among other unpredictable variables. These are elements that historical data, no matter how comprehensive, may not fully encapsulate.

To mitigate these limitations and bolster the precision of forecasts, it is advisable to introduce additional external factors into the forecasting models. These factors can encompass a broad spectrum, ranging from economic indicators like inflation rates and GDP growth to demographic trends, market sentiment indicators, and industry-specific data. By broadening the scope of variables considered, models become more adaptable and robust in the face of evolving circumstances. The inclusion of these external factors empowers the models to navigate the complexities of a dynamic environment.

Ultimately, the primary aim is to enhance the accuracy of forecasts. Historical data undeniably serves as an invaluable foundation for comprehending trends and patterns. However, augmenting this foundation with external factors equips the models with the flexibility to adjust to changing conditions and, in turn, offer predictions that are more dependable. In summary, while historical data provides valuable historical context, it is essential to acknowledge its limitations and advocate for the incorporation of external factors to refine the quality and reliability of financial forecasts. This approach better equips decision-makers to navigate the uncertainties inherent in predicting future outcomes.

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Replace '80105' with the column name corresponding to your time series data
ts = time_series_df

# Check Stationarity
def test_stationarity(timeseries):
    # Perform Augmented Dickey-Fuller test
    result = adfuller(timeseries)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:', result[4])
    
    # Plot rolling statistics
    rolling_mean = timeseries.rolling(window=12).mean()
    rolling_std = timeseries.rolling(window=12).std()
    
    plt.figure(figsize=(12, 6))
    plt.plot(timeseries, label='Original')
    plt.plot(rolling_mean, label='Rolling Mean')
    plt.plot(rolling_std, label='Rolling Std')
    plt.legend()
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()

# Differencing (if needed)
d = 1  # Order of differencing
ts_diff = ts - ts.shift(d)
ts_diff = ts_diff.dropna()

# Check Stationarity of Differenced Data
test_stationarity(ts_diff)

# ACF and PACF Plots
plt.figure(figsize=(12, 6))
plot_acf(ts_diff, lags=40, ax=plt.gca())
plt.title('ACF Plot')
plt.show()

plt.figure(figsize=(12, 6))
plot_pacf(ts_diff, lags=40, ax=plt.gca())
plt.title('PACF Plot')
plt.show()

# Fit ARIMA Model
p = 1  # Order of AR component
q = 1  # Order of MA component
model = sm.tsa.ARIMA(ts, order=(p, d, q))
results = model.fit()

# Print Model Summary
print(results.summary())

# Plot Residuals
residuals = results.resid
plt.figure(figsize=(12, 6))
plt.plot(residuals)
plt.title('Residuals')
plt.show()
```

```python
# Define the forecast horizon (number of periods to forecast)
forecast_horizon = 120  # 10 years * 12 months/year

# Create an empty DataFrame to store the forecasts
forecast_df = pd.DataFrame(columns=['RegionName', 'Date', 'Forecast'])

# Create a list to store the average returns for each RegionName
average_returns = []

# Iterate through each RegionName (ZipCode) in the top_20_filtered_zipcodes DataFrame
for zipcode in top_20_filtered_zipcodes['RegionName']:
    # Extract the time series data for the current ZipCode
    zipcode_data = time_series_df

    # Ensure the "Date" column is in datetime format
    zipcode_data['Date'] = pd.to_datetime(zipcode_data['Date'])

    # Generate future dates for forecasting
    future_dates = pd.date_range(start=zipcode_data['Date'].max(), periods=forecast_horizon + 1, freq='M')[1:]

    # Make forecasts for the next 10 years for the current ZipCode
    forecast = best_model_fit.get_forecast(steps=forecast_horizon)
    forecast_mean = forecast.predicted_mean

    # Create a DataFrame for the forecasts for the current ZipCode
    forecast_zipcode_df = pd.DataFrame({
        'RegionName': [zipcode] * forecast_horizon,
        'Date': future_dates,
        'Forecast': forecast_mean
    })

    # Append the forecasts for this RegionName (ZipCode) to the main forecast DataFrame
    forecast_df = pd.concat([forecast_df, forecast_zipcode_df])

    # Find the closest year to 2018 in the data and use its value to calculate the average return
    closest_year_2018 = zipcode_data['Year'].min()  # Initialize with the first available year
    for year in zipcode_data['Year']:
        if abs(year - 2018) < abs(closest_year_2018 - 2018):
            closest_year_2018 = year

    # Get the value for the closest year to 2018
    last_year_2018_data = zipcode_data[zipcode_data['Year'] == closest_year_2018]['Value'].values

    if len(last_year_2018_data) > 0:
        last_month_2018 = last_year_2018_data[-1]  # Use the last available value in 2018
        average_return = (forecast_mean - last_month_2018) / 10
    else:
        average_return = None

    average_returns.append((zipcode, average_return))

# Reset the index of the forecast DataFrame
forecast_df.reset_index(drop=True, inplace=True)

# Print the first few rows of the forecast DataFrame
forecast_df.head(100)

# Sort the average returns by the return value and get the top 5
top_5_average_returns = sorted(average_returns, key=lambda x: x[1], reverse=True)[:5]
print("Top 5 ZipCodes with Highest Average Returns:")
for zipcode, average_return in top_5_average_returns:
    print(f"RegionName: {zipcode}, Average Return: {average_return}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the time period for which you want to obtain predictions and confidence intervals
start_date = '1996-05-01'
end_date = '2020-04-01'

# Get predictions and confidence intervals for the specified period
pred = model_fit.get_prediction(start=pd.to_datetime(start_date), end=pd.to_datetime(end_date), dynamic=False)
pred_conf = pred.conf_int()

# Extract the predicted values
predictions = pred.predicted_mean

# Create a DataFrame to store the values
result_df = pd.DataFrame({'Predicted': predictions, 'Lower CI': pred_conf.iloc[:, 0], 'Upper CI': pred_conf.iloc[:, 1]})

# Filter the DataFrame for the desired date range (1996 to 2018)
result_df = result_df['1996':'2018']

# Set figure size
plt.rcParams['figure.figsize'] = (18, 8)

# Plot observed values
ax = result_df.plot(label='Observed')

# Plot predicted values in blue
predictions.plot(ax=ax, label='Predicted', color='blue', alpha=0.9)

# Plot the range for confidence intervals
ax.fill_between(result_df.index,
                result_df['Lower CI'],
                result_df['Upper CI'], color='lightpink', alpha=0.5, label='Confidence Interval')

# Set axes labels and title
ax.set_xlabel('Date', fontsize=20)
ax.set_ylabel('Price', fontsize=20)
ax.set_title('Testing Forecasting Model Performance', fontsize=30)

# Add legend
plt.legend(fontsize=15)

# Show the plot
plt.show()
```
