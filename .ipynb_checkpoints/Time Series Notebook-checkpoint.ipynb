### ZIP Code Selection

The assumption is that the client has an unlimited budget and  is open to real estate investments anywhere in the country, indicating a lack of specific preferences. This approach allows for a comprehensive analysis, considering a wide range of opportunities and maximizing returns.

Thus, I decided to use evaluation criteria from prior research, enhancing the analysis by leveraging existing domain knowledge, saving time, and increasing credibility. This approach ensures a well-informed and reliable evaluation process.Based on these assumptiona and the domain knowledge I chose the following metrics to refine my data towards building a robust model:

**a. Historical Price Appreciation:**
   - Calculating historical price appreciation is essential for evaluating real estate investments. It offers insights into ZIP code performance over time, considering long-term trends, risk assessment, and profit potential.

**b. SizeRank:**
   - SizeRank measures the relative size or popularity of ZIP codes. It's valuable for assessing market significance, diversification opportunities, and growth potential in real estate investments.

**c. Coefficient of Variation (CV) for each ZIP code:**
   - The CV assesses data volatility, helping in risk assessment and investment strategy. It ensures portfolio diversity by considering different market stability levels.

**d. Return on Investment (ROI):**
   - In my real estate investment analysis, the strategic use of ROI stands as a pivotal metric. ROI quantifies the profitability of each ZIP code investment by assessing the change in median housing prices from April 1996 to April 2018. This strategic choice empowers me to gauge the performance of different ZIP codes, distinguishing those with substantial property value appreciation. By employing ROI, I rank and identify investment opportunities, guiding my selection of ZIP codes for optimal returns. Thus, predicting ROI is a metric upon which my final investment opportunities' selection will be based.


**a. Historical Price Appreciation**

```python
# Select relevant columns for historical price appreciation calculation
price_columns = zillow_data.columns[7:]  # Columns from 1996-04 to 2018-04
zillow_data['Historical_Price_Appreciation'] = zillow_data[price_columns].pct_change(axis=1).mean(axis=1)

# Sort the ZIP codes by historical price appreciation in descending order
zillow_data.sort_values(by='Historical_Price_Appreciation', ascending=False).head(20)
```

**b. Size Rank**

Calculating the threshold for the top 5% of SizeRank values is a strategic approach to pinpoint markets with significant size and potential for real estate investments. It aligns with the goals of diversification, risk mitigation, and maximizing profit potential.

```python
# Calculate the threshold for the top 5% of SizeRank values
size_rank_threshold = zillow_data['SizeRank'].quantile(0.95)
size_rank_threshold
```

Close to 13986 met this criteria

**C.  Coefficient of Variation (CV) for each ZIP code**

```python
# Calculate the coefficient of variation (CV) for each ZIP code
zillow_data['CV'] = zillow_data[price_columns].std(axis=1) / zillow_data[price_columns].mean(axis=1)

# Select the top 20 ZIP codes with the highest CV
top_20_cv_zipcodes = zillow_data.nlargest(20, 'CV')
top_20_cv_zipcodes
```

In my real estate investment analysis, I have strategically decided to calculate the 60th percentile of the Coefficient of Variation (CV). This choice aligns with my risk assessment approach, catering to diverse investment strategies. By focusing on the 60th percentile, I strike a balance between stability and market variability, making it an attractive option for risk-aware investors. Additionally, this approach promotes portfolio diversity, acknowledging the value of including markets with different levels of price stability. Ultimately, my goal is to target ZIP codes that offer a balanced combination of risk and profit potential, ensuring a well-rounded and strategic investment strategy.

```python
# Calculate the 60th percentile of CV
cv_threshold = zillow_data['CV'].quantile(0.6)
```

```python
# Filter ZIP codes to meet all criteria
filtered_zipcodes = zillow_data[
    (zillow_data['Historical_Price_Appreciation'] >= zillow_data['Historical_Price_Appreciation'].quantile(0.8)) &
    (zillow_data['SizeRank'] >= size_rank_threshold) &
    (zillow_data['CV'] < cv_threshold)
]
```

**d). Return on Investment (ROI)**

```python
# Drop rows with NaN values in the columns needed for ROI calculation
filtered_zipcodes = filtered_zipcodes.dropna(subset=['1996-04', '2018-04'])

# Calculate ROI for each ZIP code based on the change in median housing prices
filtered_zipcodes['ROI'] = (filtered_zipcodes['2018-04'] - filtered_zipcodes['1996-04']) / filtered_zipcodes['1996-04']

# Select the top 10 ZIP codes with the highest ROI
top_10_filtered_zipcodes = filtered_zipcodes.nlargest(10, 'ROI')

# Display the top 10 ZIP codes that meet all criteria
print("Top 10 ZIP Codes Based on Combined Criteria:")
top_10_filtered_zipcodes[['RegionName', 'Historical_Price_Appreciation','CV', 'SizeRank', 'ROI']]
```

## 4. Data Preprocessing

In the critical data preprocessing step, I proactively addressed potential issues that could affect the time series modeling. I focused on outlier detection, handling missing values, and evaluating feature relevance. Robust statistical methods were used to identify and manage outliers. Missing data was handled using tailored strategies, and non-relevant features were removed. This meticulous data curation ensured dataset quality for my subsequent modeling efforts.


```python
def get_datetimes(top_10_filtered_zipcodes):
    """
    Takes a dataframe:
    returns only those column names that can be converted into datetime objects 
    as datetime objects.
    """
    date_columns = [col for col in top_10_filtered_zipcodes.columns[7:] if re.match(r'\d{4}-\d{2}', col)]  # Filter columns with "YYYY-MM" format
    return pd.to_datetime(date_columns, format='%Y-%m')

# Use the modified get_datetimes function to obtain datetime columns
datetime_columns = get_datetimes(top_10_filtered_zipcodes)
datetime_columns
```

```python
top_10_filtered_zipcodes.dropna(axis=1)
```

```python
# Function to identify and remove outliers using Z-score
# Function to remove outliers using Z-score
def remove_outliers(data, column_name, threshold=3):
    # Convert the column to numeric (if it's not already)
    data[column_name] = pd.to_numeric(data[column_name], errors='coerce')
    
    # Calculate Z-scores for numeric values
    z_scores = np.abs(stats.zscore(data[column_name].dropna()))
    
    # Create a mask for outliers
    is_outlier = z_scores <= threshold
    
    # Return the DataFrame without outliers
    return data[is_outlier]

# Exclude the specified columns
exclude_columns = ['RegionID', 'RegionName', 'City', 'State', 'Metro', 'CountyName']

# Get the list of numeric columns excluding the specified columns
numeric_columns = [col for col in top_10_filtered_zipcodes.columns if col not in exclude_columns]

# Remove outliers for each numeric column and update the DataFrame
for column_name in numeric_columns:
    top_10_filtered_zipcodes = remove_outliers(top_10_filtered_zipcodes, column_name)

# Now, the 'top_10_filtered_zipcodes' DataFrame has outliers removed for each numeric column
```

```python
# Check for missing values in the DataFrame
missing_values = top_10_filtered_zipcodes.isnull().sum()

# List columns with missing values (if any)
columns_with_missing_values = missing_values[missing_values > 0]

# Display the columns with missing values and the number of missing values
if not columns_with_missing_values.empty:
    print("Columns with missing values:")
    print(columns_with_missing_values)
else:
    print("No missing values found in the DataFrame.")
```

## 6.Time Series Modelling

### a) ZIP Code Selection

In time series modeling, I utilize historical data to make predictions about the future. To initiate the Time Series model, my initial step is to pinpoint a specific zipcode within the country, where I aim to predict future ROI. My strategy involves selecting a single zipcode to construct my Time Series model initially, and for this purpose, I've opted for the one exhibiting the most substantial growth since 2011, namely, ZipCode-38461. This chosen zipcode will serve as my starting point for forecasting its future value. Subsequently, I will employ a similar workflow on other selected zipcodes, systematically identifying the top 5 zipcodes anticipated to yield the highest ROI.

```python
#Dataframe containining historical prices of the 38461 Zip Code.

zip38461 = top_10_filtered_zipcodes[top_10_filtered_zipcodes['RegionName']==38461]
zip38461.head()
```

```python
#Meltinga and reshaping a DataFrame while converting date columns to datetime format, removing rows with missing values in the 'value' column, and returning the resulting DataFrame.

def melt_data(zip38461):
    melted = pd.melt(zip38461 , id_vars=['RegionID','RegionName','CountyName','CV','Historical_Price_Appreciation','City','State','Metro','SizeRank','ROI'], 
                     var_name='Date')
    melted['Date'] = pd.to_datetime(melted['Date'])
    melted = melted.dropna(subset=['value'])
    return melted

melted_df = melt_data(zip38461)
melted_df
```

```python
# Convert the 'Date' column to a datetime type
melted_df['Date'] = pd.to_datetime(melted_df['Date'])

# Set the 'Date' column as the index
time_series_df = melted_df.set_index('Date')

# Select the 'value' column as the time series data
time_series_df = time_series_df[['value']]
time_series_df
```

```python
#Visualizing the Time Series Data for the zipcode

fig, ax = plt.subplots(figsize=(20,12))
ax.plot(time_series_df)

ax.set_xlabel('Year', fontsize=30)
ax.set_ylabel('Price (in 100,000($)',fontsize=20)
ax.set_title('Price History for Zip Code 38461 ',fontsize=30);
```

The pattern observed in the graph is quite similar to what was encountered in the previous exploration of other zip codes.


Its observed that the market conditions before the 2010s do not necessarily mirror the current market situation. Hence, I have decided to focus the Time Series modeling solely on data from 2011 onwards.

```python
recent_series = time_series_df['2011':]
```

```python
#Visualizing Time Series Data for zipcode- 38461 from 2011

fig, ax = plt.subplots(figsize=(20,12))
ax.plot(recent_series)

ax.set_xlabel('Year', fontsize=20)
ax.set_ylabel('Price in 100,000($)',fontsize=20)
ax.set_title('Zip Code 38461 Price (since 2011)',fontsize=30);
```

Observation:There is  a favorable trend characterized by a moderately consistent linear increase in average prices.

### b) Modelling

```python
# Splitting the last 24 months of the series as a test dataset,representing approximately 29% of the data.
train_series = recent_series[:'2016-04']
test_series = recent_series['2016-05':]
```

```python
# Selecting the model

auto_model = pm.auto_arima(train_series, start_p=0, start_q=0,
                     test='adf',
                     max_p=5, max_q=5,
                     m=1,
                     d=0,
                     seasonal = True,
                     start_P=0,start_Q=0,
                     D=0,
                     trace=True,
                     error_action= 'ignore',
                     suppress_warnings=True,
                     stepwise=True,with_intercept=False)

print(auto_model.summary())
```

This code automatically selects the best ARIMA model for the time series and provides a summary of the selected model's parameters and statistics.I shall proceed to plug in the optimal parameter values precscribed

```python
# Plug the optimal parameter values into a new SARIMAX model
ARIMA_MODEL = sm.tsa.statespace.SARIMAX(train_series, 
                                        order=(3, 0, 2), 
                                        seasonal_order=(0, 0, 0, 0), 
                                        enforce_stationarity=False, 
                                        enforce_invertibility=False)

# Fit the model and print results
output = ARIMA_MODEL.fit()

print(output.summary().tables[1])
auto_model.plot_diagnostics(figsize=(18, 18))
plt.show()
```

Inspecting the diagnostic plots is crucial to ensure that my residuals exhibit no correlation and adhere to a normal distribution. If these assumptions are not met, I may need to refine my model.

a) The residuals closely align with the standard normal distribution (N(0,1)), indicating normal distribution characteristics.

b) The QQ-plot demonstrates that the residuals adhere to a linear trend line, confirming their normal distribution.

c) The correlogram plot illustrates minimal correlations between time series residuals and their lagged counterparts. This suggests the absence of evident seasonality in my data.

These observations instill confidence that my model does not have correlations and effectively fits my data for future price forecasting.


### c) Testing Forecasting Model Performance


Obtaining the predictions generated by my model for the time spanning from May 2016 to April 2018, which aligns with the duration used for my test dataset. I can visualize these predictions on a shared chart to assess their proximity or divergence from one another.

```python
# Get predictions starting from 04-01-2015 and calculate confidence intervals
pred = output.get_prediction(start=pd.to_datetime('2015-05'), end=pd.to_datetime('2018-04'), dynamic=False)
pred_conf = pred.conf_int()
pred_conf
```

**Plotting this  against the test data**

```python
# Set the figure size using plt.rcParams
plt.rcParams['figure.figsize'] = (18, 8)

# Plot observed values
ax = recent_series.plot(label='observed')

# Plot predicted values
pred.predicted_mean.plot(ax=ax, label='Prediction Series', alpha=0.9)

# Plot the range for confidence intervals
ax.fill_between(pred_conf.index,
                pred_conf.iloc[:, 0],
                pred_conf.iloc[:, 1], color='lightpink', alpha=0.5, label='Confidence Interval')

# Set axes labels
ax.set_xlabel('Date', fontsize=20)
ax.set_ylabel('Price', fontsize=20)
ax.set_title('Testing Forecasting Model Performance(Zip Code 38461) ', fontsize=30)
plt.legend()
plt.show()
```

While the model is not flawless in its accuracy, it did successfully capture the general trend that unfolded between May 2015 and April 2018. With confidence in its performance over this 2-3 year timeframe, I can now proceed to predict future trends until April 2023. This endpoint corresponds to a full three years from the purchase date in April 2018

## 7.Forecasting 


In this section, I delve into the critical aspect of forecasting. Leveraging the insights gained from my model's performance evaluation and its ability to capture historical trends, I now shift my focus toward predicting future developments and outcomes. My objective is to provide forecasts and projections that extend beyond the current point in time to capture the best 5 ZIP Codes to invest in from the list of 10.

```python
# Plug the parameter values from  Auto ARIMA model into a new SARIMAX model that fits the entire series.
ARIMA_MODEL = sm.tsa.statespace.SARIMAX(recent_series, 
                                        order=(3,0,2), 
                                        seasonal_order=(0, 0, 0, 0), 
                                        enforce_stationarity=False, 
                                        enforce_invertibility=False)

# Fit the model and print results
full_output = ARIMA_MODEL.fit()

print(output.summary().tables[1])
```

```python
## Getting a forecast for the next 84 months after the last recorded date on the dataset.
forecast = full_output.get_forecast(84)
future_prediction = forecast.conf_int()
future_prediction['value'] = forecast.predicted_mean
future_prediction.columns = ['lower','upper','prediction'] 
future_prediction
```

Having generated an eight-year forecast, it's time to create a visual representation to gauge the projected growth of the 38461 zipcode region by April 2025

```python
#Plotting the forecast

fig, ax = plt.subplots()
recent_series.plot(ax=ax,label='Real Values')


future_prediction['prediction'].plot(ax=ax,label='predicted value',ls='--')

ax.fill_between(x= future_prediction.index, y1= future_prediction['lower'], 
                y2= future_prediction['upper'],color='lightpink',
                label='Confidence Interval')
ax.legend() 
plt.ylabel("Average Price")
plt.title('Average Home Price - 38461 - With Forcasted Value & Confidence Intervals')
plt.show()
```

Now that the methodology for predicting future prices in zipcode 38461 is outlined, I can seamlessly integrate this entire process into a repetitive loop to obtain my anticipated prices in April 2025 for the other selected filtered zipcodes.

```python
def melt_data(top_10_filtered_zipcodes):
    melted = pd.melt(top_10_filtered_zipcodes , id_vars=['RegionID','RegionName','CountyName','CV','Historical_Price_Appreciation','City','State','Metro','SizeRank','ROI'], 
                     var_name='Date')
    melted['Date'] = pd.to_datetime(melted['Date'])
    melted = melted.dropna(subset=['value'])
    return melted

melted_df1 = melt_data(top_10_filtered_zipcodes)
melted_df1
```

```python
# Convert the 'Date' column to a datetime type
melted_df1['Date'] = pd.to_datetime(melted_df1['Date'])

# Set the 'Date' column as the index
time_series_df1 = melted_df1.set_index('Date')

# Select the 'value' column as the time series data
time_series_df1 = time_series_df1[['value']]
time_series_df1
```

```python
zip_predictions = {}

# Loop through the top 10 filtered zipcodes
for zipcode in top_10_filtered_zipcodes['RegionName']:
    
  # Melting the time series of the given zipcode
    series1 = melt_data(top_10_filtered_zipcodes[top_10_filtered_zipcodes['RegionName'] == zipcode])
    
    # Convert the 'Date' column to a datetime type and set it as the index
    series1['Date'] = pd.to_datetime(series1['Date'])
    series1.set_index('Date', inplace=True)
    
    # Only taking data from 2013 onwards to more accurately reflect current market conditions
    recent_series = series1['2013':]
    
    # Splitting the last 24 months of my series as a test dataset.
    train_series = recent_series[:'2016-04']
    test_series = recent_series['2016-05':]
    
    
    # Auto ARIMA model
    auto_model = pm.auto_arima(train_series['value'], start_p=0, start_q=0,
                     test='adf',
                     max_p=5, max_q=5,
                     m=1,
                     d=0,
                     seasonal=True,
                     start_P=0, start_Q=0,
                     D=0,
                     trace=True,
                     error_action='ignore',
                     suppress_warnings=True,
                     stepwise=True, with_intercept=False)
   
    # Plug the optimal parameter values for my Training data into a SARIMAX model that fits my entire series.
    ARIMA_MODEL = sm.tsa.statespace.SARIMAX(recent_series['value'], 
                                            order=auto_model.order, 
                                            seasonal_order=auto_model.seasonal_order, 
                                            enforce_stationarity=False, 
                                            enforce_invertibility=False)

    # Fit the model and print results
    output = ARIMA_MODEL.fit()

    # Getting a forecast for the next 84 months after the last recorded date on the dataset.
    forecast = output.get_forecast(steps=84)
    prediction = forecast.conf_int()
    prediction['value'] = forecast.predicted_mean
    prediction.columns = ['lower', 'upper', 'prediction'] 
    
    # Adding the Zipcode's ROI to the zip_predictions dictionary
    zip_predictions[zipcode] = ((prediction['prediction'][-1])
                                - (recent_series['value'][-1])) / (recent_series['value'][-1])
```

## 8. Recommendations and Conclusion

```python
# Sorting the 8-year ROI forecast for zipcodes into descending order
sort_orders = sorted(zip_predictions.items(), key=lambda x: x[1], reverse=True)

sorted_forecast_8yr = {}
for i in sort_orders:
    sorted_forecast_8yr[i[0]] = i[1]

# Selecting only the Top 5 Zips
top_5_zipcodes = list(sorted_forecast_8yr.items())[:5]

fig, ax = plt.subplots(figsize=(18, 12))

x_labels = [top_5_zipcodes[0][0], top_5_zipcodes[1][0], top_5_zipcodes[2][0],
            top_5_zipcodes[3][0], top_5_zipcodes[4][0]]
x = [1, 2, 3, 4, 5]
y = [top_5_zipcodes[0][1], top_5_zipcodes[1][1], top_5_zipcodes[2][1],
     top_5_zipcodes[3][1], top_5_zipcodes[4][1]]

ax.bar(x, y)
remaining_zipcodes = list(sorted_forecast_8yr.items())[5:]
median_ROI_other_zips = np.median([roi for _, roi in remaining_zipcodes])

ax.axhline(y=median_ROI_other_zips, label='Median ROI for remaining Zips')
ax.set_xticks(x)
ax.set_xticklabels(x_labels)
ax.set_yticks([(a/100) for a in list(range(0, 201, 25))])  # Adjusted y-ticks range to 0-200%
ax.set_yticklabels([str(a)+'%' for a in list(range(0, 201, 25))])  # Adjusted y-tick labels
ax.set_title('Top 5 Recommendations', fontsize=20)
ax.set_ylabel('Average ROI', fontsize=20)
ax.set_xlabel('Zipcodes', fontsize=20)
ax.legend(prop={'size': 15});
```

The reliability of predictions in financial forecasting heavily relies on historical monthly returns, forming the foundation of the analysis. Investments and economic conditions are inherently subject to uncertainties and risks beyond what historical data can capture. These uncertainties may include abrupt policy changes, technological advancements, market sentiment shifts, and global events, among other variables.

To enhance forecast accuracy, it's advisable to incorporate additional external factors into forecasting models. These factors can range from economic indicators to demographic trends, market sentiment, and industry-specific data. By considering a broader set of variables, models become more adaptable and robust in the face of changing circumstances.

The ultimate goal is to improve forecast accuracy by combining historical data with external factors. While historical data provides valuable context, recognizing its limitations and advocating for the inclusion of external factors can refine the quality and reliability of financial forecasts. This approach equips decision-makers to better navigate the uncertainties inherent in predicting future outcomes.
